```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
from pylab import rcParams
import seaborn as sb
from numpy.random import randn
import csv
from datetime import date, datetime

%matplotlib inline
```

```python

file1 = 'C:/Users/Wayne/Desktop/First Data Project/Mass Shooting/Mass.csv'
file2 = 'C:/Users/Wayne/Desktop/First Data Project/Mass Shooting/Mass1.csv'
file3 = 'C:/Users/Wayne/Desktop/First Data Project/Mass Shooting/Mass1.xls'

df = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_excel(file3)
```

```python
#del df2['Unnamed: 0']
df3['Date'] = pd.to_datetime(df3.Date)
df3['DOW'] = df3['Date'].dt.weekday_name
df3 = df3.set_index('Date')
```

```python
#plt.scatter(df2['Location'], df2['Fatalities'])
#plt.plot(df2['State'], marker='.', linestyle = 'none')
#plt.plot(df3['Region'], marker='.',markersize=12, linestyle = 'none')
fig, ax = plt.subplots(facecolor='lightgray')
ax.axis([0, 600, 0, 10])

sb.stripplot(x="Total victims",y="DOW", data=df3, jitter=True, linewidth=1)
plt.title('Day of the Week of Mass Shooting 1982-2018')
ax.text(0.6467,0.5 , "Vegas Shooting", transform=ax.transAxes)

plt.xlabel("Number of Victims Shot Per Case")
plt.ylabel("Day of the Week")
plt.xticks(rotation=45)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
from pylab import rcParams
import seaborn as sb
from numpy.random import randn
import csv
from datetime import date, datetime

%matplotlib inline
```

```python

file3 = 'C:/Users/Wayne/Desktop/First Data Project/Mass Shooting/Mass1.xls'
df3 = pd.read_excel(file3)
```

```python
#del df2['Unnamed: 0']
df3['Date'] = pd.to_datetime(df3.Date)
df3['DOW'] = df3['Date'].dt.weekday_name
df3 = df3.set_index('Date')
```

```python
#plt.scatter(df2['Location'], df2['Fatalities'])
#plt.plot(df2['State'], marker='.', linestyle = 'none')
#plt.plot(df3['Region'], marker='.',markersize=12, linestyle = 'none')
fig, ax = plt.subplots(figsize=(10,6),facecolor='lightgray')
ax.annotate('Vegas Shooting: 547 Shot',size=20,xy=(530,2.7),xycoords='data', xytext=(-60, 40),textcoords='offset points',ha='right',arrowprops=dict(facecolor='black'))

ax.axis([0, 600, 0, 10])

#ax.text(0.6467,0.54 , "Vegas Shooting 547 Shot", transform=ax.transAxes)
sb.stripplot(x="Total victims",size=12,y="DOW", data=df3, jitter=True, linewidth=1)
plt.title('Day of the Week of Mass Shooting 1982-2018')
plt.xlabel("Number of Victims: Shot Per Case")
plt.ylabel("Day of the Week")
plt.xticks(rotation=45)
```

```python
#need to create a "ledgend" that will anotate the population percents
#of each region

plt.plot(df3['Region'], marker='.',markersize=12, linestyle = 'none')
plt.title('Number of Mass Shootings by US Regions')
plt.xlabel('Year of Case')
plt.ylabel('Region of the United States')
plt.xticks(rotation = 90)

plt.show()
```
